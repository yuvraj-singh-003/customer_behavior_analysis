As you can see, there are 37 null values in Review Rating Column, Let's fix it !

```python
new_df = df;
```

Now, we will fill the null

#### ðŸ‘‰ Now we will fill the null values with the median of each category to prevent biasness in our data

```python
new_df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median))
```

```python
df = new_df
```

#### Lets Check the null values Again

```python
df.isnull().sum()
```

#### ðŸ‘‰ In this we convert the column name into camel case format for better understanding and accessing

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_")
df = df.rename(columns={"purchase_amount_(usd)": "purchase_amount"})
```

### ðŸ“Œ Feature Engineering

#### ðŸ‘‰ Now we will create age labels based on age group

```python
labels =  ["Young Adult", "Adult", "Middle Aged", "Senior"]

df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age', 'age_group']].head(10)
```

#### ðŸ‘‰ We also create frequency mapping (numeric) for purchase frequence days (names)

```python
frequency_mapping = {
    "Fortnightly" : 14,
    "Weekly" : 7,
    "Monthly" : 30,
    "Quarterly" : 90,
    "Bi-Weekly" : 14,
    "Annually" : 365,
    "Every 3 Months": 90
}

df["purchase_frequency_days"] = df["frequency_of_purchases"].map(frequency_mapping)


```

```python
df[["frequency_of_purchases", "purchase_frequency_days"]].head(5)
```
